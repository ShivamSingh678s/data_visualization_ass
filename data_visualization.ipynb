```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

n_samples = 100
X1 = np.random.randn(n_samples)
X2 = 2 * X1 + np.random.randn(n_samples) * 0.5
X3 = -1.5 * X1 + np.random.randn(n_samples) * 0.3
X4 = np.random.randn(n_samples)

df = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3,
    'Feature 4': X4
})

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# np.random.seed(20)
# data = {
#     'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#     'Day': np.random.choice(range(1, 31), 100),
#     'Sales': np.random.randint(1000, 5000, 100)
# }
# df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

heatmap_fig = px.imshow(sales_pivot, labels=dict(color="Sales"),
                        title="Heatmap of Sales Across Days and Months", aspect='auto')

heatmap_fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x^2 + y^2))
import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

surface_fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

surface_fig.update_layout(
    title='3D Surface Plot of z = sin((x + y))',
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    ),
    width=800, height=600
)

surface_fig.show()
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(10, 100, size=n)
colors = [f'#{np.random.randint(0, 0xFFFFFF):06x}' for _ in range(n)]

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

source = ColumnDataSource(data)

p = figure(title="Bokeh Scatter Plot with Random Data", x_axis_label='X', y_axis_label='Y')

p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='sizes', fill_alpha=0.6)
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

# Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data)

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits")

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.legend.title = "Fruits"

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

x = edges[:-1]
y = hist

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Count')

p.quad(top=y, bottom=0, left=x, right=edges[1:], fill_color='blue', line_color='black', alpha=0.7)

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=400, height=400, tools='')

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=mapper, line_color='white')

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')
show(p)
```
